```
from imageai.Prediction.Custom import ModelTraining

model_trainer = ModelTraining()
model_trainer.setModelTypeAsResNet()
model_trainer.setDataDirectory("/content/gdrive/My Drive/Colab Notebooks/flowers")
model_trainer.trainModel(num_objects=5, num_experiments=100, enhance_data=False, batch_size=32, show_network_summary=True)
```

```
# Show Image
import cv2
import numpy as np
from matplotlib import pyplot as plt

def show_image(image_name):
  img = cv2.imread(image_name)
  img_cvt=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow(img_cvt)
  plt.show()
```

```
from imageai.Prediction.Custom import CustomImagePrediction
import os

execution_path = os.getcwd()

#image_path = "/content/daisy.jpg"
image_path = "/content/roses.jpg"

prediction = CustomImagePrediction()
prediction.setModelTypeAsResNet()
prediction.setModelPath(os.path.join(execution_path, "/content/gdrive/My Drive/Colab Notebooks/flowers/models/model_ex-100_acc-0.781250.h5"))
prediction.setJsonPath(os.path.join(execution_path, "/content/gdrive/My Drive/Colab Notebooks/flowers/json/model_class.json"))
prediction.loadModel(num_objects=5)

predictions, probabilities = prediction.predictImage(os.path.join(execution_path, image_path), result_count=5)

# Display Image
show_image(image_path)

for eachPrediction, eachProbability in zip(predictions, probabilities):
    print(eachPrediction , " : " , eachProbability)
```
